```python
lines = []
with open('geo_dataset.csv', 'r') as f:
    header = f.readline()
    lines.append(header)
    for l in f.readlines():
        entry = l.split(';')
        date = entry[3]
        d,h = date.split()
        j,m,a = d.split('-')
        entry[3] = "%s-%s-%s %s"%(a,m,j,h)
        lines.append(';'.join(entry))
    
```

```python
with open('newgeo.csv','w') as f:
    for l in lines:
        f.write(l+'\n')
```
