```python
def build_dic(hexstr):
    dic = [0]*16
    for char in hexstr:
        dic[int(char, 16)] += 1
    return sorted(dic)
```

```python
from tqdm import tqdm
from sympy.utilities.iterables import multiset_permutations
p = ['um', 'urhs', 'tye', 'lqx', 'gz', 'niz', 'jx', 'tpf', 'x', 'elrt', 'pt', 'ku', 'kwb', 'bz', 'f', 'h'] #, 'x', 'tye', 'tpf', 'um', 'niz', 'lrt', 'ku', 'wb', 'bz']
for i in tqdm(range(16)):
    for j in range(i, 16):
        for k in range(j, 16):
            for l in range(k, 16):
                for m in range(l, 16):
                    for n in range(m, 16):
                        for a, b, c, d, e, f in multiset_permutations([i, j, k, l, m, n]):
                            g2, h2, i2, j2, k2, l2 = p[a], p[b], p[c], p[d], p[e], p[f]
                            if sha256(g2 + h2 + i2 + j2 + k2 + l2) == privi:
                                print(g2 + h2 + i2 + j2 + k2 + l2)
```
